# **Señales continuas** 

```python
# importar librerias 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# parte dos configraciones iniciales 
w = 1
fx = lambda t: np.sin(w*t)

t0 = 0
tn = 2*np.pi
muestras = 51

#parte 3 procedimiento 
ti = np.linspace(t0, tn, muestras)
senal = fx(ti)
```

```python
#Parte 4- salida y gráfica 
np.set_printoptions(precision=4)
print("tiempo: ")
print(ti)
print("señal: x(t)")
print(senal)

#grafica
plt.figure()
plt.plot(ti, senal)
plt.xlabel('t[s]')
plt.ylabel('Señal [x(t)]')
plt.show
```

# **Señal Discreta**

```python
w = 2*np.pi / 12
fx = lambda n: np.sin(w*n) 
```

```python
# Parte3 - procedimiento 
#vector de n muestras 
n0 = 0
muestras = 12 + 1

ni = np.arange(n0, n0+muestras, 1)
senal_discreta = fx(ni)
 
```

```python
# Salida y Gráfica 
 
np.set_printoptions(precision=4)
print("n: ")
print(ni)
print("señal: x(n)")
print(senal_discreta)

#grafica
plt.figure()
plt.stem(ni, senal_discreta)
plt.xlabel('n[s]')
plt.ylabel('Señal discreta S[x(n)]')
plt.show
```

# **Suceciones Unitarias**

```python
#Señal 1 - Impulso unitario 
x1 = np.array([0,0,0,1,0,0,0,0])
n1 = np.arange(-4 , 4)

plt.figure()
plt.stem(n1, x1)
plt.xlabel('n')
plt.ylabel('Ampllitud')
plt.title('Impulso unitario')
plt.show()
```

```python
x2 = np.concatenate((np.zeros(10), np.ones(11)))
n2 = np.arange(-10 , 11)

plt.figure()
plt.stem(n2, x2)
plt.xlabel('n')
plt.ylabel('Ampllitud')
plt.title('Impulso escalon')
plt.show()
```

```python
a = 0.8 
n3 = np.arange(-5,21)
x3 = np.concatenate((np.zeros(5), a**np.arange(21)))

plt.figure()
plt.stem(n3, x3)
plt.xlabel('n')
plt.title('exponencial real')
plt.show()
```
