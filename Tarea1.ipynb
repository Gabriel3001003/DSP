```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Rango de valores de n
n = np.arange(0, 300)

# Definicin de las seales
x_a = np.cos(0.01 * np.pi * n)  # Peridica con N = 200
x_b = np.cos(np.pi * (30/105) * n)  # Peridica con N = 7
x_c = np.cos(3 * np.pi * n)  # No peridica
x_d = np.sin(3 * n)  # No peridica
x_e = np.sin(np.pi * (62/10) * n)  # No peridica

# Crear figuras
plt.figure(figsize=(12, 10))

# Subgrfica para cada seal
plt.subplot(3, 2, 1)
plt.stem(n, x_a)
plt.title(r'$\cos(0.01\pi n)$, Peridica (N=200)')
plt.xlim([0, 250])
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(3, 2, 2)
plt.stem(n, x_b)
plt.title(r'$\cos\left(\pi \frac{30}{105} n \right)$, Peridica (N=7)')
plt.xlim([0, 50])
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(3, 2, 3)
plt.stem(n, x_c)
plt.title(r'$\cos(3\pi n)$, No peridica')
plt.xlim([0, 50])
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(3, 2, 4)
plt.stem(n, x_d)
plt.title(r'$\sin(3n)$, No peridica')
plt.xlim([0, 50])
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(3, 2, 5)
plt.stem(n, x_e)
plt.title(r'$\sin\left(\pi \frac{62}{10} n \right)$, No peridica')
plt.xlim([0, 50])
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(0, 50, 1)  # Para seales discretas
T = np.linspace(0, 10, 1000)  # Para seales continuas

# Seal  x_a(t) = 3 cos(5t + /6)
x_a = 3 * np.cos(5 * T + np.pi / 6)

# Seal  x(n) = 3 cos(5n + /6)
x_b = 3 * np.cos(5 * n + np.pi / 6)

# Seal  x(n) = 2 |exp(j(n/6 - ))|
x_c = 2 * np.ones_like(n)  # La seal es constante en 2

# Sea x(n) = cos(n/8) cos(n/8)
x_d = np.cos(n / 8) * np.cos(np.pi * n / 8)

# Seal  x(n) = cos(n/2) - sin(n/8) + 3cos(n/4 + /3)
x_e = np.cos(np.pi * n / 2) - np.sin(np.pi * n / 8) + 3 * np.cos(np.pi * n / 4 + np.pi / 3)

# Graficamos
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
fig.suptitle("Grficas de las seales del ejercicio")

# Grficas
axs[0, 0].plot(T, x_a, label='x_a(t) = 3cos(5t + /6)')
axs[0, 0].set_title("(a) Seal Continua")
axs[0, 0].grid()
axs[0, 0].legend()

axs[0, 1].stem(n, x_b, label='x(n) = 3cos(5n + /6)')
axs[0, 1].set_title("(b) Seal Discreta")
axs[0, 1].grid()
axs[0, 1].legend()

axs[1, 0].stem(n, x_c, label='x(n) = 2 |exp(j(n/6 - ))|')
axs[1, 0].set_title("(c) Seal Discreta (Constante)")
axs[1, 0].grid()
axs[1, 0].legend()

axs[1, 1].stem(n, x_d, label='x(n) = cos(n/8)cos(n/8)')
axs[1, 1].set_title("(d) Seal Discreta")
axs[1, 1].grid()
axs[1, 1].legend()

axs[2, 0].stem(n, x_e, label='x(n) = cos(n/2) - sin(n/8) + 3cos(n/4 + /3)')
axs[2, 0].set_title("(e) Seal Discreta")
axs[2, 0].grid()
axs[2, 0].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# (a) Seal continua x_a(t) = 3 sin(100t)
t = np.linspace(0, 0.03, 1000)  # de 0 a 30ms
y_a = 3 * np.sin(100 * np.pi * t)

plt.figure(figsize=(10, 4))
plt.plot(t * 1000, y_a, label=r'$x_a(t) = 3\sin(100\pi t)$')
plt.xlabel("Tiempo (ms)")
plt.ylabel("Amplitud")
plt.title("(a) Seal Continua")
plt.grid()
plt.legend()
plt.show()

# (b) Muestreo con Fs = 300 Hz
Fs = 300  # Frecuencia de muestreo en Hz
Ts = 1 / Fs  # Perodo de muestreo
n = np.arange(0, 18)  # Varias muestras para visualizar periodicidad
t_n = n * Ts  # Instantes de muestreo
x_n = 3 * np.sin((np.pi / 3) * n)  # Seal discreta x(n)

# Mostrar la periodicidad
plt.figure(figsize=(10, 4))
plt.stem(n, x_n, markerfmt='ro', basefmt=" ", linefmt='r-', label=r'$x(n) = 3\sin(\frac{\pi}{3}n)$')
plt.xlabel("Muestras (n)")
plt.ylabel("Amplitud")
plt.title("(b) Seal Discreta - Periodicidad con N = 6")
plt.grid()
plt.legend()
plt.show()

# (c) Calcular y mostrar valores en un perodo
data_values = [(n_i, 3 * np.sin((np.pi / 3) * n_i)) for n_i in range(6)]
print("Valores en un periodo de x(n):")
for n_i, x_val in data_values:
    print(f"n = {n_i}, x(n) = {x_val:.2f}")

# (d) Clculo de Fs mnimo para que x(n) alcance 3
Fs_min = 100  # Fs mnimo para alcanzar el mximo de la seal
print(f"(d) La frecuencia mnima de muestreo Fs para captar la amplitud mxima es {Fs_min} Hz")
```

# **EJERCICIO 1.7*-*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definicin de la seal continua
Fs = 600  # Frecuencia de muestreo en Hz
T = 1 / Fs  # Perodo de muestreo
t = np.linspace(0, 10 / Fs, 1000)  # 10 perodos de muestreo para visualizacin
xa_t = np.sin(480 * np.pi * t) + 3 * np.sin(720 * np.pi * t)

# (a) Frecuencia de Nyquist
f1 = 240  # Frecuencia en Hz
f2 = 360  # Frecuencia en Hz
Nyquist_rate = 2 * max(f1, f2)
print(f"(a) La frecuencia de Nyquist es {Nyquist_rate} Hz")

# (b) Frecuencia de plegamiento
Folding_freq = Fs / 2
print(f"(b) La frecuencia de plegamiento es {Folding_freq} Hz")

# (c) Frecuencias en el dominio discreto
omega_1 = (2 * np.pi * f1) / Fs  # Frecuencia en radianes
omega_2 = (2 * np.pi * f2) / Fs  # Frecuencia en radianes
print(f"(c) Frecuencia en radianes: 1 = {omega_1:.2f}, 2 = {omega_2:.2f}")

# Seal discreta
n = np.arange(20)  # Primeros 20 valores de muestra
t_n = n * T
x_n = np.sin(omega_1 * n) + 3 * np.sin(omega_2 * n)

# Grfica de la seal continua
plt.figure(figsize=(10, 4))
plt.plot(t, xa_t, label='Seal continua $x_a(t)$', alpha=0.7)
plt.scatter(t_n, x_n, color='red', label='Muestras $x(n)$')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("(d) Seal Continua y Muestras Discretas")
plt.legend()
plt.grid()
plt.show()
```

# **Ejercicio 1.11**

```python

f0 = 50                                 # Frecuencia ms baja en Hz de la seal
f1 = 125                                # Frecuencia ms alta en Hz de la seal  
COS0 = lambda t: 3*np.cos(2*np.pi*f0*t)
SIN0 = lambda t: 2*np.sin(2*np.pi*f1*t)

Fs = 200                                 # Frecuencia de muestreo en Hz
Ts = 1/Fs                                # Periodo de muestreo en segundos

ti = np.arange(0, 2/f0, 1/(f1*100))      # vector de tiempo continuo
ni = np.arange(0, 2/f0, Ts)              # vector de tiempo discreto

signal_a = COS0(ti) + SIN0(ti)           # Seal analgica
signal_d = COS0(ni) + SIN0(ni)           # Seal muestreada

# Grfica:
plt.figure()    
plt.plot(ti, signal_a, label='Seal Analgica', color='red')
plt.stem(ni, signal_d, label='Seal Discreta', basefmt=" ")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

# **Ejercicio 1.15**

```python
F0 = [0.5, 2, 3, 4500]                         # Frecuencia de la seal
Fs = 5000                                      # Frecuencia de muestreo

ni = np.arange(0, 100, 1)                       # vector de tiempo discreto

# Grfica:
def plot_signal(F0):
    Xn = lambda n: np.sin(2*np.pi*(F0/Fs)*n)
    Xn_values = Xn(ni)  # Seal muestreada

    # Grfica:
    plt.figure()
    plt.stem(ni, Xn_values, label=f'Seal Discreta F0={F0}Hz', basefmt=" ")
    plt.xlabel('n')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid(True)
    plt.show()

    return Xn_values

# Plot signals for each frequency in F0
for f in F0:
    plot_signal(f)
```

```python
F0 = 2000  # Frecuencia de la seal en Hz
Fs = 50000  # Frecuencia de muestreo en Hz

Xn = plot_signal(F0) #1-

# 2-
Yn = Xn[::2]
yi = np.arange(0, len(Yn), 1)  # vector de tiempo discreto

# Grfica:
plt.figure()
plt.stem(yi, Yn, label='Seal Discreta Submuestreada', basefmt=" ")
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
